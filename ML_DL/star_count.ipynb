## In this exercise, we are going to count no. of stars in the sky 

```python
from matplotlib import pyplot as plt
from skimage import data
from skimage.feature import blob_dog ,blob_log,blob_doh
from math import sqrt
from skimage.color import rgb2gray
import glob
from skimage.io import imread
```

```python
file=glob.glob(r"C:\Users\prem\Desktop\ML_DL\star_image1.jpg")[0]
im=imread(file,as_grey=True,extent=[0,.5,0,.5])
plt.imshow(im,cmap='gray')
plt.show()
```

```python
blobs_log = blob_log(im, max_sigma=30, num_sigma=10, threshold=.1)
# Compute radii in the 3rd column.
blobs_log[:, 2] = blobs_log[:, 2] * sqrt(2)
numrows = len(blobs_log)
print("Number of stars counted : " ,numrows)
```

### Lets mark bounded circle for every star that we have counted

```python
fig, ax = plt.subplots(1, 1)
plt.imshow(im, cmap='gray')
for blob in blobs_log:
    y, x, r = blob
    c = plt.Circle((x, y), r+5, color='lime', linewidth=.5, fill=False)
    ax.add_patch(c)
```
