```python
import numpy as np

#Input array
X=np.array([[1,0,1,0],[1,0,1,1],[0,1,0,1]]) # here, our i/p X is a data matrix of size 3*4
print("shape of input = ", X.shape)

#Output
y=np.array([[1],[1],[0]]) # output y takes size of 3*1

# defining our own Sigmoid Function
def sigmoid (x):
    return 1/(1 + np.exp(-x))

# Derivative of Sigmoid Function to update gradients while back propogating
def derivatives_sigmoid(x):
    return x * (1 - x) # this is derivative of sigmoid function

#Variable initialization
epoch=5000 #Setting training iterations
lr=0.1 #Setting our own learning rate
inputlayer_neurons = X.shape[1] #number of features in data set(X.Shape[1] returns 4 in (3,4) tuple)
hiddenlayer_neurons = 3 #number of hidden layers neurons (pre-defined)
output_neurons = 1 #number of neurons at output layer (pre-defined, should be 1 for simple architechture)

#weight and bias initialization

wh=np.random.uniform(size=(inputlayer_neurons,hiddenlayer_neurons))
# o/p should be a 4*3 matrix with random numbers b/w 0&1.here, we intialize weights b/w 1st & 2nd layers

bh=np.random.uniform(size=(1,hiddenlayer_neurons))
# o/p should be a 1*3 matrix with random no's b/w 0&1.here, we intialize bias term of i/p layer

wout=np.random.uniform(size=(hiddenlayer_neurons,output_neurons))
# o/p should be a 3*1 matrix with random no's b/w 0&1.here, we intialize weights b/w 2nd & o/p layer

bout=np.random.uniform(size=(1,output_neurons))
# o/p should be a 1*1 matrix with random no's b/w 0&1.here, we intialize bias term of hidden layer

print("weights intialized between i/p and hidden layers :\n",wh)
print("bias intialized at hidden layer :\n", bh)
print("weights intialized between hidden and o/p layer :\n",wout)
print("bias intialized at o/p layer :\n",bout)

for i in range(epoch): # 1 FORWARD + 1 BACKWARD propogation = 1 EPOCH 

    #Forward Propogation
    hidden_layer_input1=np.dot(X,wh) #dot product b/w input matrix(3*4) and weights of i/p layer(4*3)=(3*3 matrix)
    hidden_layer_input=hidden_layer_input1 + bh #just adding a scalar-value of bias term
    hiddenlayer_activations = sigmoid(hidden_layer_input) # applying sigmoid non-linear function to (3*3 matrix)
    output_layer_input1=np.dot(hiddenlayer_activations,wout) # dot-pro b/w prev.o/p & present weights(result=3*1)
    output_layer_input= output_layer_input1+ bout # addition of hidden-layer bias term(3*1)
    output = sigmoid(output_layer_input) # applying sigmoid-activation to o/p (3*1)matrix
    print("output we got at epoch ",i," \n",output) # SHOULD BE 3*1 MATRIX
    print("===================================================================================")

    #Backpropagation
    E = y-output #defining our error term (3*1)matrix--->diff. b/w defined o/p and o/p we got (ERROR)
    slope_output_layer = derivatives_sigmoid(output) # calculating derivative at o/p layer
    slope_hidden_layer = derivatives_sigmoid(hiddenlayer_activations) #derivative at hidden-layer
    d_output = E * slope_output_layer # error * (derivative we calculated at o/p layer)(3*1 matrix)
    #print("*******************************",slope_output_layer.shape,"****************************")
    Error_at_hidden_layer = d_output.dot(wout.T) # a dot product b/w (3*1) and (1*3) matrix (chain-rule)
    d_hiddenlayer = Error_at_hidden_layer * slope_hidden_layer # calculating delta of hidden layer(3*3 matrix)
    wout += hiddenlayer_activations.T.dot(d_output) *lr #(update o/p weights using prev. derivatives and lr)
    bout += np.sum(d_output, axis=0,keepdims=True) *lr#(update bias terms with sum of prev. partial derivatives)
    wh += X.T.dot(d_hiddenlayer) *lr # update hidden weights using input matrix and delta of hidden layer
    bh += np.sum(d_hiddenlayer, axis=0,keepdims=True) *lr # update hidden bias term

print("final output, \n",output)
```
